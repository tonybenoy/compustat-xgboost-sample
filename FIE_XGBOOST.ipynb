Mean absolute error (MAE) is a measure of the average magnitude of the errors in a set of predictions, without taking their direction into account. It is calculated as the sum of the absolute errors divided by the number of observations.

MAE is a useful metric for evaluating the performance of regression models because it is easy to understand and interpret. It is also a relatively robust metric, meaning that it is not as sensitive to outliers as some other metrics, such as mean squared error (MSE).

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test, y_pred)
```

R-squared (R²) is a statistical measure of how well a regression model explains the variation in the dependent variable. It is calculated as the proportion of the variation in the dependent variable that is predictable from the independent variables.

R-squared is a useful metric for evaluating the performance of regression models because it is easy to understand and interpret. It ranges from 0 to 1, with a higher value indicating a better fit.

```python
from sklearn.metrics import r2_score
```
